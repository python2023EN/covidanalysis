```python
# Join the DataFrames
covid = pd.merge(cov, vac, how='left', on=['Province/State','Date'])
```

```python
# Explore the new DataFrame by looking at shape, types of data and head
covid.head()
```

```python
covid.shape
```

```python
## Explore the new DataFrame by looking at types of data
covid.dtypes
```

```python
# Fix the date column data type
covid['Date'] = pd.to_datetime(covid['Date'],format='%Y-%m-%d')
```

```python
#Drop unnecessary columns 
covid = covid.drop(columns=['Country/Region_y',
               'Lat_y',
               'Long_y',
               'ISO 3166-1 Alpha 3-Codes_y',
               'Sub-region Name_y', 
               'Intermediate Region Code_y'])
```

```python
#Rename columns
covid = covid.rename(columns=
                    {'Country/Region_x':'Country/Region',
                    'Lat_x':'Lat',
                    'Long_x':'Long',
                    'ISO 3166-1 Alpha 3-Codes_x':'ISO 3166-1 Alpha 3-Codes', 
                    'Sub-region Name_x':'Sub-region Name',
                    'Intermediate Region Code_x':'Intermediate Region Code'})
covid.head()
```

```python
#Looking for missing values
covid_na = covid[covid.isna().any(axis=1)]
covid_na
```

```python
#Fill with bfill, since the values tend to rise
covid['Deaths'] = covid['Deaths'].fillna(method='bfill')
covid['Cases'] = covid['Recovered'].fillna(method='bfill')
covid['Recovered'] = covid['Recovered'].fillna(method='bfill')
covid['Hospitalised'] = covid['Hospitalised'].fillna(method='bfill')
```

```python
#Verify that missing values have been replaced
covid_na = covid.isna().sum()
covid_na
```

```python
#Change the data type to int 
covid['Deaths'] = covid['Deaths'].astype(int)
covid['Cases'] = covid['Cases'].astype(int)
covid['Recovered'] = covid['Recovered'].astype(int)
covid['Hospitalised'] = covid['Hospitalised'].astype(int)
```

```python
#Check types in covid
covid.dtypes
```

```python
covid = pd.DataFrame(covid)
```

```python
#Group by Province/State 
groupbystate = covid.groupby(['Province/State']
                            )[['Vaccinated', 'First Dose', 'Second Dose']].sum()
#Show the values sorted by Descending order by number of Vaccinated
groupbystate.sort_values('Vaccinated', ascending=False)
```

```python
#Create new column Difference 
groupbystate['Difference'] = groupbystate['First Dose'] - groupbystate['Second Dose']
#Show values sorted by Descending order by number of not fully vaccinated
groupbystate.sort_values('Difference', ascending = False)
```

```python
#We can see that Gibraltatr and Monserat Have the highest number of people
#with First dose but not second in absolute terms
```

```python
#Calculate ratio of people eligible for second dose but do not have it
groupbystate['Ratio'] = groupbystate['Difference'] / groupbystate['First Dose'] * 100
#Show values sorted by Descending order by RATIO
groupbystate.sort_values('Ratio', ascending = False)
```

```python
#From here we can see that only around 4.51% of people did not get a second dose
```

```python
#Calculate First Percentage to show people who are vaccinated out of all eligigle
groupbystate['First Percentage'] = groupbystate['Second Dose'] / groupbystate['First Dose'] * 100
#Show values sorted in ascending order of First Percentage
groupbystate.sort_values('First Percentage')
```

```python
#Clearly around 95.5% of people eligible for a second dose are vaccinated
```

```python
groupbystate['Province/State'] = groupbystate.index
```

```python
#Create graph showing number of people without second dose
sns.set_theme(style="whitegrid")
ax = sns.barplot(x = 'Difference', y = 'Province/State', data = groupbystate\
                .sort_values('Difference'), palette='Blues_d')
ax.set_xlabel('People with 1st but not second')
ax.set_ylabel('Province/State')
ax.set_title('People with 1st vaccination but not 2nd by Province/State')
plt.savefig('peopleunvacabsolute')
```

```python
sns.set_theme(style="whitegrid")
ax2 = sns.barplot(x = 'First Percentage', y = 'Province/State', data = groupbystate\
                .sort_values('First Percentage'), palette='pastel')
ax2.set_xlabel('Percentage of 1st vaccination out of total')
ax2.set_ylabel('Province/State')
ax2.set_title('Percentage fully vaccinated out of total eligible by Province/State')
plt.savefig('fullyvacpercentage.png')
```
